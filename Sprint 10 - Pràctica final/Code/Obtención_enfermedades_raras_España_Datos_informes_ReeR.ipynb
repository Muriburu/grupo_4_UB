# **Informe situación enfermedades raras en España - Ministerio de Salud**

#### Extraer tablas según página

```python
import pdfplumber
import pandas as pd
import os

def extraer_y_guardar_tablas_pdf(ruta_pdf, carpeta_base=r"C:\Users\maria\ITAcademy\Sprint 10 - Pràctica final\Rare Diseases\tablas_extraidas_informes_ReeR"):
    # Extraer nombre base del archivo PDF (sin extensión)
    nombre_pdf = os.path.splitext(os.path.basename(ruta_pdf))[0]
    carpeta_salida = os.path.join(carpeta_base, f"tablas_{nombre_pdf}")
    os.makedirs(carpeta_salida, exist_ok=True)
    total_guardadas = 0
    with pdfplumber.open(ruta_pdf) as pdf:
        for i, page in enumerate(pdf.pages):
            tablas = page.extract_tables()
            if tablas:
                for j, tabla in enumerate(tablas):
                    df = pd.DataFrame(tabla[1:], columns=tabla[0])
                    nombre_archivo = f"tabla_p{i+1}_t{j+1}.csv"
                    ruta_completa = os.path.join(carpeta_salida, nombre_archivo)
                    df.to_csv(ruta_completa, index=False)
                    total_guardadas += 1
    return carpeta_salida, total_guardadas
```

## **1. Informe ReeR 2021 - Situación de las Enfermedades Raras en 2010‐2018**

##### Leer el pdf

```python
with pdfplumber.open(r"C:\Users\maria\ITAcademy\Sprint 10 - Pràctica final\Rare Diseases\Bibliografía RD\Ministerio de Salud - Situación de las enfermedades raras en España\Informe_reer_reducido_2021.pdf") as pdf:
    for i, page in enumerate(pdf.pages):
        text = page.extract_text()
        print(f"--- Página {i+1} ---")
        print(text) 
```

##### Extraer tablas

```python
# Uso de la función
ruta_pdf = r"C:\Users\maria\ITAcademy\Sprint 10 - Pràctica final\Rare Diseases\Bibliografía RD\Ministerio de Salud - Situación de las enfermedades raras en España\Informe_reer_reducido_2021.pdf"
carpeta, total = extraer_y_guardar_tablas_pdf(ruta_pdf)
print(f"Carpeta creada: {carpeta}")
print(f"Total de archivos CSV generados: {total}")
```
